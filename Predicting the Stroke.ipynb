```python
import pandas as pd
df = pd.read_csv('newData.csv')
df.head()
```

```python
inputs = df.drop(['stroke','id'],axis='columns')
tareget = df['stroke']
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_gender = LabelEncoder()
le_hypertension = LabelEncoder()
le_heart_disease =  LabelEncoder()
le_ever_married = LabelEncoder()
le_work_type= LabelEncoder()
le_Residence_type = LabelEncoder()
le_smoking_status =  LabelEncoder()
```

```python
inputs['gender_n'] = le_gender.fit_transform(inputs['gender'])
inputs['age_n'] = inputs['age']
inputs['hypertension_n'] = le_hypertension.fit_transform(inputs['hypertension'])
inputs['heart_disease_n'] = le_heart_disease.fit_transform(inputs['heart_disease'])
inputs['ever_married_n'] = le_ever_married.fit_transform(inputs['ever_married'])
inputs['work_type_n'] = le_work_type.fit_transform(inputs['work_type'])
inputs['Residence_type_n'] = le_Residence_type.fit_transform(inputs['Residence_type'])
inputs['avg_glucose_level_n'] = inputs['avg_glucose_level']
inputs['bmi_n'] = inputs['bmi']
inputs['smoking_status_n'] = le_smoking_status.fit_transform(inputs['smoking_status'])
inputs.head()
```

```python
inputs_n = inputs.drop(['gender','hypertension','Residence_type','heart_disease','ever_married','work_type','smoking_status','bmi','avg_glucose_level','age'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,tareget)
```

```python
model.score(inputs_n,tareget)
```
